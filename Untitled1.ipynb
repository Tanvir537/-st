```python
# Step 1: Upload the image
from google.colab import files
uploaded = files.upload()

# Extract uploaded filename
import os
filename = list(uploaded.keys())[0]

# Step 2: Import required libraries
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

# Step 3: Load the image in grayscale
img = cv.imread(filename, cv.IMREAD_GRAYSCALE)
assert img is not None, "Image could not be loaded. Check the filename."

# Step 4: Apply vertical then horizontal flip
v_flip = cv.flip(img, 0)         # Vertical flip
vh_flip = cv.flip(v_flip, 1)     # Then horizontal flip
cv.imwrite("flip_vh.png", vh_flip)

# Step 5: Apply rotation on the first output (vh_flip)
(h, w) = vh_flip.shape
center = (w // 2, h // 2)
rotation_matrix = cv.getRotationMatrix2D(center, 45, 1.0)  # 45 degree rotation
rotated = cv.warpAffine(vh_flip, rotation_matrix, (w, h))
cv.imwrite("rotated.png", rotated)

# Step 6: Apply 30% zoom on rotated image
zoom_factor = 1.3
zoomed = cv.resize(rotated, None, fx=zoom_factor, fy=zoom_factor)
zh, zw = zoomed.shape

# Crop back to original size centered (assume tumor is in center)
start_x = zh//2 - h//2
start_y = zw//2 - w//2
zoom_cropped = zoomed[start_x:start_x+h, start_y:start_y+w]
cv.imwrite("zoomed.png", zoom_cropped)

# Step 7: Show all images
titles = ["Original", "Flip VH", "Rotated", "Zoomed"]
images = [img, vh_flip, rotated, zoom_cropped]

plt.figure(figsize=(12, 8))
for i in range(4):
    plt.subplot(2, 2, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()

!git config --global user.name "tanvir537"
!git config --global user.email "tr60780@gmail.com"
!git clone https://github.com/Tanvir537/Resarch.git

!cp flip_vh.png rotated.png zoomed.png Resarch.git/
%cd Resarch.git
!git add .
!git commit -m "Added transformed MRI images"
!git pus
```
